```python
from pymongo import MongoClient
import pandas as pd
```

```python
client = MongoClient()
db = client.yelp
business_df = pd.DataFrame(list(db.business.find()))
```

```python
drop_columns = ['neighborhoods', 'type', 'full_address', 'hours', 'state','attributes', '_id', 'open']
business_df.drop(labels = drop_columns, axis = 1)
```

```python
business_df.to_csv('Business.csv', encoding = 'utf-8')
```
